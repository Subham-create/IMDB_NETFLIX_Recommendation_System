### Problem Statement 2
### If a producer wants to release some content, which month must he do so?( Month when least amount of content is added)

```python
## Before we solve this problem let's visualize the ratio of TV shows and Movies in a pie chart
pie=df2.type.value_counts().reset_index()
pie=pie.rename(columns={'type': 'count','index':'type'})
pie
```

```python
fig_pie=go.Figure(data=[go.Pie(labels=pie['type'], values=pie['count'],hole=0.3)],layout=go.Layout(title="Distribution of Type of Content"))
```

```python
## As we need to know th eMonth when least amount of content is added we need to split the date column and do some feature engineering
df2.info()
```

we can see that date is stored as string let's split into month and year


```python
df2=df2.dropna() ## Let's drop all the Nans
```

```python
time=pd.DataFrame()
```

```python
time['yr']=df2['date_added'].apply(lambda x:x.split(", ")[-1])
time['daymon']=df2['date_added'].apply(lambda x:x.split(", ")[0])
time['month']=time['daymon'].apply(lambda x: x.lstrip().split(' ')[0])
time=time.drop("daymon",axis=1)
```

```python
time['yr'].value_counts()
```

```python
time['month'].value_counts()
```

```python
content = go.Figure(data=[go.Histogram2d(
        x=time.yr,
        y=time.month,
    )],layout=go.Layout(title="Content added per year with distribution in month"))
        
```

There are three posibilities on releasing content on Netflix:- 
- If we are considering just year 2019 then " SEPTEMBER" looks the best year
- If we are considering overall then "JUNE" looks the best ( But we will rule out this one as most of the content were added in 2018 and 2019.
- If we consider the average of just 2018 and 2019 then it would be "JUNE" with only 184 Release, next would be "MAY" with 186 release

### PROBLEM STATEMENT 3
### Which tv series has the highest number of seasons? *

```python
df2.head()
```

```python
tv=df2[(df2.type=='TV Show')]
tv.head()
```

```python
tv.columns
```

```python
tv=tv.drop(['show_id', 'type', 'country', 'date_added', 'release_year',
       'rating', 'listed_in', 'description'],axis=1)
```

```python
tv.head()
```

```python
tv['duration']=tv['duration'].apply(lambda x:x.split(" ")[0])
tv.info()
```

```python
tv.info()
```

```python
tv=tv.sort_values(by='seasons',ascending=True)
tv.tail(10)
```

```python
import plotly.figure_factory as ff

tvshow=go.Figure([go.Bar(x=tv['title'].tail(20),y=tv['seasons'].tail(20))],layout=go.Layout(title="TV Shows with Maximum Seasons"))
```

```python
tab=ff.create_table(tv.sort_values(by='seasons',ascending=False).head(20))
```

### Problem Statement 5
### Name the country that has the second highest rated content. (Use imdb Ratings dataset, imdb movies dataset and Netflix dataset) ?

```python
## We can use the same dataset to answer this question
top_movie.info()
```

```python
rated=pd.DataFrame(top_movie.country.value_counts())
```

```python
rated=rated.reset_index()
rated=rated.rename(columns={'index':'country','country':'count'})
```

```python
rated.head()
```

```python
country=go.Figure(go.Treemap(labels=rated['country'],values=rated['count'],parents=["","", "", "", "", "", "", "", "", "",]),layout=go.Layout(title="Top 10 Countries with highest rated content"))
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html

app = dash.Dash()
app.layout = html.Div([
    dcc.Graph(figure=fig),
    dcc.Graph(figure=fig_pie),
    dcc.Graph(figure=content),
    dcc.Graph(figure=tvshow),
    dcc.Graph(figure=tab),
    dcc.Graph(figure=rate),
    dcc.Graph(figure=tpm),
    dcc.Graph(figure=country),
])

app.run_server(debug=True, use_reloader=False)
```
